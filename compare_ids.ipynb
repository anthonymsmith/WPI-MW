```python

import pandas as pd

# Load the data
default_clustering_file = '1213Clustering - LatestMetrics.xlsx'
clustering_file = input("Enter the name of the clustering data file:") or default_clustering_file

default_patrons_file = 'Patrons.csv'
patrons_file = input("Enter the name of the patrons data file:") or default_patrons_file

clustering_data = pd.read_excel(clustering_file, sheet_name='Sheet1')
print(f'read {clustering_file}')
patrons_data = pd.read_csv('patrons_file')
print(f'read {patrons_file}\n')

# Extract identifiers
clustering_contact_ids = set(clustering_data['ContactId'].dropna())
patrons_account_ids = set(patrons_data['AccountId'].dropna())

# Identify overlaps and discrepancies
common_ids = clustering_contact_ids.intersection(patrons_account_ids)
clustering_only_ids = clustering_contact_ids.difference(patrons_account_ids)
patrons_only_ids = patrons_account_ids.difference(clustering_contact_ids)

# Analyze duplicates in clustering data
clustering_duplicates = clustering_data[clustering_data.duplicated(subset=['ContactId'], keep=False)]
print(f'read {patrons_file}\n')

# Prepare detailed outputs for clustering-only and patrons-only
clustering_only_df = clustering_data[clustering_data['ContactId'].isin(clustering_only_ids)][['ContactId', 'Cluster KMedoids', 'Segment']]
patrons_only_df = patrons_data[patrons_data['AccountId'].isin(patrons_only_ids)][['AccountId', 'Segment']]

# Save the results to CSV files
clustering_only_df.to_csv('clustering_only_ids.csv', index=False)
patrons_only_df.to_csv('patrons_only_ids.csv', index=False)
print('wrote output files.\n')

# Results summary
print("Summary of Comparison:")
print(f"Total Clustering Entries: {len(clustering_data)}")
print(f"Duplicate Entries in Clustering: {len(clustering_duplicates)}")
print(f"Unique Clustering ContactIds: {len(clustering_contact_ids)}")
print(f"Unique Patrons AccountIds: {len(patrons_account_ids)}")
print(f"Common IDs: {len(common_ids)}")
print(f"Clustering-Only IDs: {len(clustering_only_ids)}")
print(f"Patrons-Only IDs: {len(patrons_only_ids)}")
print("\nOutput files created: 'clustering_only_ids.csv' and 'patrons_only_ids.csv'")
"""
import pandas as pd

# Load the data
clustering_data = pd.read_excel('1213Clustering - LatestMetrics.xlsx', sheet_name='Sheet1')
patrons_data = pd.read_csv('Patrons.csv')

# 1. Count of ContactIds by Cluster KMedoids and Segment
clustering_counts = (
    clustering_data.groupby(['Cluster KMedoids', 'Segment'])
    .size()
    .reset_index(name='ContactId_Count')
)

# 2. Identify unassigned ContactIds in clustering data (missing Cluster or Segment)
unassigned_contact_ids = clustering_data[
    clustering_data['Cluster KMedoids'].isna() | clustering_data['Segment'].isna()
    ]['ContactId'].dropna()

# 3. Count of AccountIds by Segment in patrons data
patrons_counts = (
    patrons_data.groupby(['Segment'])
    .size()
    .reset_index(name='AccountId_Count')
)

# Save results to files for review
clustering_counts.to_csv('clustering_counts.csv', index=False)
unassigned_contact_ids.to_csv('unassigned_contact_ids.csv', index=False, header=['Unassigned_ContactId'])
patrons_counts.to_csv('patrons_counts.csv', index=False)

# Print summary
print("Summary of Results:")
print("\nClustering Counts (by Cluster KMedoids and Segment):")
print(clustering_counts)
print("\nUnassigned ContactIds in Clustering File:")
print(unassigned_contact_ids)
print("\nPatrons Counts (by Segment):")
print(patrons_counts)

print("\nOutput files created:")
print("- 'clustering_counts.csv': Count of ContactIds by KMedoids and Segment")
print("- 'unassigned_contact_ids.csv': List of unassigned ContactIds")
print("- 'patrons_counts.csv': Count of AccountIds by Segment")
"""
```
